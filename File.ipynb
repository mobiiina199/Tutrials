```python
def process_file(filename):
    with open(filename, 'r') as fh:
        
        for line in fh:
            line = line.rstrip('\n')
            if len(line) > 0:
                if line[0] == '#':
                    return
        
            if len(line) > 1:
                if line[0:2] =='//':
                    return
                
            print(line)
```

```python
def main():
    try:
        with open('E:/data/data science/data course/colors.txt') as fh:
            colors = []
            for line in fh:
                colors.append(line.rstrip('\n'))
    except IOError:
        print('could not open colors.txt')
        exit()
    
    for i in range(len(colors)):
        print('{} {}'.format(i, colors[i]))
        
    c = int(input('select color: '))
    print(colors[c])
    
main()
        
```
